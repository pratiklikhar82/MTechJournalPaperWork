```python
count = 1
for user in users:
    image_dir = os.path.join(data_dir,user,'0','binary')
    image_list = os.listdir(image_dir)
    for image in image_list:
        im = cv2.imread(os.path.join(image_dir,image))
        Full_frame_mask=np.zeros([480,640,3],dtype="uint8")
        Full_frame_mask[190:400,190:500] = im
        ht = Full_frame_mask.shape[0]
        wd = Full_frame_mask.shape[1]
        for i in range(ht):
            for j in range(wd):
                if Full_frame_mask[i][j][0] != 0:
                    Full_frame_mask[i][j][0] = 3
                    Full_frame_mask[i][j][1] = 3
                    Full_frame_mask[i][j][2] = 3
                else:
                    Full_frame_mask[i][j][0] = 0
                    Full_frame_mask[i][j][1] = 0
                    Full_frame_mask[i][j][2] = 0
        if count >= 1 and count <= 9:
            cv2.imwrite(os.path.join(root_dir,'modified_dataset','masks')+'/0_000'+str(count)+'.png', Full_frame_mask)
        elif count >= 10 and count <= 99:
            cv2.imwrite(os.path.join(root_dir,'modified_dataset','masks')+'/0_00'+str(count)+'.png', Full_frame_mask)
        elif count >= 100 and count <= 999:
            cv2.imwrite(os.path.join(root_dir,'modified_dataset','masks')+'/0_0'+str(count)+'.png', Full_frame_mask)
        elif count >= 1000:
            cv2.imwrite(os.path.join(root_dir,'modified_dataset','masks')+'/0_'+str(count)+'.png', Full_frame_mask)
        print('image {} processed'.format(count))
        count += 1
    print('User {} processed'.format(user))
```
